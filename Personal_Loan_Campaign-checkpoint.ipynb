## Exploratory Data Analysis (EDA)

EDA helps us understand the distribution of individual variables and relationships 
between variables before building any model. We perform both Univariate Analysis 
(one variable at a time) and Bivariate Analysis (two variables together, especially 
against our target variable Personal_Loan).

```python
# ============================================================
# CELL 4a: EDA HELPER FUNCTIONS
# ============================================================

# --- FUNCTION 1: histogram_boxplot ---
# Used for continuous numerical columns (Age, Income, Mortgage, CCAvg, Experience)
# Combines a boxplot on top and histogram below in one chart
# Boxplot shows: median, IQR, and outliers as dots
# Histogram shows: the actual distribution shape
# Green dashed line = Mean, Black solid line = Median
def histogram_boxplot(data, feature, figsize=(12, 7), kde=False, bins=None):
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,
        sharex=True,
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )
    sns.boxplot(data=data, x=feature, ax=ax_box2, showmeans=True, color="violet")
    sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins, palette="winter") \
        if bins else sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2)
    ax_hist2.axvline(data[feature].mean(), color="green", linestyle="--")
    ax_hist2.axvline(data[feature].median(), color="black", linestyle="-")
    plt.show()


# --- FUNCTION 2: labeled_barplot ---
# Used for categorical columns (Education, Family, CreditCard, Online, etc.)
# Shows a bar chart with the count OR percentage labeled on top of each bar
# perc=True shows percentages, perc=False shows raw counts
def labeled_barplot(data, feature, perc=False, n=None):
    total = len(data[feature])
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))
    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n].sort_values(),
    )
    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(100 * p.get_height() / total)
        else:
            label = p.get_height()
        x = p.get_x() + p.get_width() / 2
        y = p.get_height()
        ax.annotate(label, (x, y), ha="center", va="center", size=12,
                    xytext=(0, 5), textcoords="offset points")
    plt.show()

print("Helper functions defined successfully!")
```

### Bivariate Analysis

Now we examine how each variable relates to Personal_Loan, our target variable. 
This helps identify which factors actually drive loan acceptance.

```python
# ============================================================
# CELL 5a: BIVARIATE HELPER FUNCTIONS
# ============================================================

# --- FUNCTION 1: stacked_barplot ---
# Used for CATEGORICAL variables vs Personal_Loan
# Shows a table of counts, then a stacked bar chart
# Each bar is split into two colors showing the proportion who took 
# the loan (1) vs didn't (0) within that category
def stacked_barplot(data, predictor, target):
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(
        by=sorter, ascending=False
    )
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(
        by=sorter, ascending=False
    )
    tab.plot(kind="bar", stacked=True, figsize=(count + 5, 5))
    plt.legend(loc="lower left", frameon=False)
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()


# --- FUNCTION 2: distribution_plot_wrt_target ---
# Used for CONTINUOUS variables vs Personal_Loan
# Shows 4 charts: distribution for loan=0, distribution for loan=1,
# boxplot comparison, and boxplot without outliers for clearer comparison
def distribution_plot_wrt_target(data, predictor, target):
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    target_uniq = data[target].unique()

    axs[0, 0].set_title("Distribution of target for target=" + str(target_uniq[0]))
    sns.histplot(data=data[data[target] == target_uniq[0]], x=predictor, kde=True,
                 ax=axs[0, 0], color="teal", stat="density")

    axs[0, 1].set_title("Distribution of target for target=" + str(target_uniq[1]))
    sns.histplot(data=data[data[target] == target_uniq[1]], x=predictor, kde=True,
                 ax=axs[0, 1], color="orange", stat="density")

    axs[1, 0].set_title("Boxplot w.r.t target")
    sns.boxplot(data=data, x=target, y=predictor, ax=axs[1, 0], palette="gist_rainbow")

    axs[1, 1].set_title("Boxplot (without outliers) w.r.t target")
    sns.boxplot(data=data, x=target, y=predictor, ax=axs[1, 1], showfliers=False, palette="gist_rainbow")

    plt.tight_layout()
    plt.show()

print("Bivariate helper functions defined successfully!")
```

## Model Building

### Model Evaluation Criterion

- We will use **Recall** as our primary evaluation metric, not Accuracy.
- Business context: missing a customer who would have accepted the loan 
  (False Negative) is more costly than wasting a marketing contact on a 
  customer who declines (False Positive). The bank loses a real revenue 
  opportunity in the first case, but only a minor outreach cost in the second.
- Accuracy alone would be misleading here since only ~9.5% of customers 
  accept the loan — a model that predicts "No" for everyone would already 
  score ~90% accuracy while being completely useless to the business.
- We will still track Accuracy, Precision, and F1 for a complete picture, 
  but Recall on the test set will be our deciding factor when comparing models.

```python
# ============================================================
# CELL 7a: MODEL EVALUATION HELPER FUNCTIONS
# ============================================================

# FUNCTION 1: Computes all 4 metrics in one go for any model
def model_performance_classification_sklearn(model, predictors, target):
    """
    model: the trained classifier
    predictors: X data (features)
    target: y data (actual labels)
    """
    # Generate predictions using the trained model
    pred = model.predict(predictors)

    # Calculate each metric by comparing actual (target) vs predicted (pred)
    acc = accuracy_score(target, pred)
    recall = recall_score(target, pred)
    precision = precision_score(target, pred)
    f1 = f1_score(target, pred)

    # Package results into a clean one-row DataFrame for easy comparison later
    df_perf = pd.DataFrame(
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1},
        index=[0],
    )
    return df_perf


# FUNCTION 2: Plots a visual confusion matrix with percentages
def confusion_matrix_sklearn(model, predictors, target):
    """
    Shows actual vs predicted outcomes as a 2x2 grid with counts and percentages
    """
    y_pred = model.predict(predictors)
    cm = confusion_matrix(target, y_pred)

    # Format each cell to show both count and percentage of total
    labels = np.asarray(
        [["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
         for item in cm.flatten()]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.show()

print("Model evaluation functions defined successfully!")
```

```python
# ============================================================
# CELL 7b: BUILDING THE BASELINE DECISION TREE MODEL
# ============================================================

# Initializing a Decision Tree Classifier with default settings (no pruning yet)
# criterion="gini" measures how "impure" a split is - the tree picks 
# splits that make each resulting group as pure (one-class-dominant) as possible
# random_state=1 ensures reproducible results every time we run this
model = DecisionTreeClassifier(criterion="gini", random_state=1)

# Training (fitting) the model on our training data
# The tree learns rules like "if Income > X and CD_Account = 1, then..."
model.fit(X_train, y_train)

print("Decision Tree model trained successfully!")
```

```python
# ============================================================
# CELL 7c: CHECKING MODEL PERFORMANCE ON TRAINING DATA
# ============================================================

# First, the confusion matrix - shows exactly how many predictions 
# were correct vs incorrect, broken down by type
confusion_matrix_sklearn(model, X_train, y_train)
```

```python
# ============================================================
# CELL 7d: TRAINING PERFORMANCE METRICS
# ============================================================

decision_tree_perf_train = model_performance_classification_sklearn(model, X_train, y_train)
print("Decision Tree Performance on Training Data:")
decision_tree_perf_train
```

## Model Performance Improvement

### Pre-Pruning

```python
# ============================================================
# CELL 8a: PRE-PRUNING - GRID SEARCH OVER PARAMETERS
# ============================================================

# Defining ranges of values to try for each pruning parameter
# np.arange(2, 7, 2) creates [2, 4, 6] - depths to try
max_depth_values = np.arange(2, 7, 2)
max_leaf_nodes_values = [50, 75, 150, 250]
min_samples_split_values = [10, 30, 50, 70]

# Variables to track the best model found so far
best_estimator = None
best_score_diff = float("inf")   # smallest gap between train/test recall
best_test_score = 0.0            # highest test recall

# Looping through every combination of the 3 parameters
# This is called a "grid search" - we're testing a grid of possibilities
for max_depth in max_depth_values:
    for max_leaf_nodes in max_leaf_nodes_values:
        for min_samples_split in min_samples_split_values:

            # Build a tree with this specific combination of settings
            # class_weight='balanced' tells the model to pay extra attention 
            # to the minority class (loan takers, only ~9.5% of data)
            estimator = DecisionTreeClassifier(
                max_depth=max_depth,
                max_leaf_nodes=max_leaf_nodes,
                min_samples_split=min_samples_split,
                class_weight="balanced",
                random_state=42,
            )

            # Train this specific tree
            estimator.fit(X_train, y_train)

            # Get predictions on both train and test sets
            y_train_pred = estimator.predict(X_train)
            y_test_pred = estimator.predict(X_test)

            # Calculate Recall for both - remember Recall is our priority metric
            train_recall_score = recall_score(y_train, y_train_pred)
            test_recall_score = recall_score(y_test, y_test_pred)

            # The gap between train and test Recall tells us how overfit the model is
            # Smaller gap = better generalization
            score_diff = abs(train_recall_score - test_recall_score)

            # Keep this combination if it has a smaller gap AND a better test score
            # than anything we've found so far
            if (score_diff < best_score_diff) & (test_recall_score > best_test_score):
                best_score_diff = score_diff
                best_test_score = test_recall_score
                best_estimator = estimator

# Print out what combination won
print("Best parameters found:")
print(f"Max depth: {best_estimator.max_depth}")
print(f"Max leaf nodes: {best_estimator.max_leaf_nodes}")
print(f"Min samples split: {best_estimator.min_samples_split}")
print(f"Best test recall score: {best_test_score}")
```

```python
# ============================================================
# CELL 8b: FITTING THE BEST PRE-PRUNED MODEL
# ============================================================

estimator = best_estimator
estimator.fit(X_train, y_train)

print("Pre-pruned model fitted.")
```

#### Note on Grid Search Approach

Our first pre-pruning attempt used `class_weight="balanced"` and selected 
the model with the smallest train/test Recall gap and highest test Recall. 
This produced a model with Recall = 1.0 but Precision = only 0.31 
(734 false positives on training data) — the model was simply predicting 
"will take loan" for nearly everyone, which technically maximizes Recall 
but is not a usable model for the business.

**Revised approach:** We removed `class_weight="balanced"` and switched 
the selection criterion to **F1 score** (the harmonic mean of Recall and 
Precision) on the test set. This prevents the grid search from rewarding 
a model that games Recall at the expense of Precision, and selects a 
tree that balances catching real loan-takers against avoiding excessive 
false positives.

```python
# ============================================================
# CELL 8a (REVISED): PRE-PRUNING - GRID SEARCH OVER PARAMETERS
# ============================================================

# Same parameter ranges as before
max_depth_values = np.arange(2, 7, 2)
max_leaf_nodes_values = [50, 75, 150, 250]
min_samples_split_values = [10, 30, 50, 70]

best_estimator = None
best_f1_test = 0.0   # we now track the BEST F1 SCORE on test data, not just recall

for max_depth in max_depth_values:
    for max_leaf_nodes in max_leaf_nodes_values:
        for min_samples_split in min_samples_split_values:

            # NOTE: removed class_weight="balanced" - it was causing the 
            # model to over-predict the minority class unrealistically
            estimator = DecisionTreeClassifier(
                max_depth=max_depth,
                max_leaf_nodes=max_leaf_nodes,
                min_samples_split=min_samples_split,
                random_state=42,
            )

            estimator.fit(X_train, y_train)
            y_test_pred = estimator.predict(X_test)

            # F1 balances Recall and Precision - prevents the 
            # "predict yes to everyone" shortcut we just saw
            test_f1 = f1_score(y_test, y_test_pred)

            # Keep the model with the highest test F1 score
            if test_f1 > best_f1_test:
                best_f1_test = test_f1
                best_estimator = estimator

print("Best parameters found:")
print(f"Max depth: {best_estimator.max_depth}")
print(f"Max leaf nodes: {best_estimator.max_leaf_nodes}")
print(f"Min samples split: {best_estimator.min_samples_split}")
print(f"Best test F1 score: {best_f1_test}")
```

```python
# ============================================================
# CELL 8b: FITTING THE BEST PRE-PRUNED MODEL
# ============================================================

estimator = best_estimator
estimator.fit(X_train, y_train)

print("Pre-pruned model fitted.")
```

### Post-Pruning

Unlike pre-pruning, post-pruning allows the tree to grow fully first, 
then prunes back branches that contribute little to overall performance, 
using a technique called Cost Complexity Pruning (CCP).

```python
# ============================================================
# CELL 9a: COMPUTING THE COST COMPLEXITY PRUNING PATH
# ============================================================

# Building a fully-grown tree (no restrictions) - same as our baseline
clf = DecisionTreeClassifier(random_state=1)

# cost_complexity_pruning_path() calculates every possible alpha value 
# and the resulting "impurity" of the tree at each pruning level
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

print(f"Number of alpha values found: {len(ccp_alphas)}")
print(f"Alpha range: {ccp_alphas.min():.5f} to {ccp_alphas.max():.5f}")
```

```python
# ============================================================
# CELL 9b: BUILDING A TREE FOR EACH ALPHA VALUE
# ============================================================

# For each alpha value, we train a separate Decision Tree
# Higher alpha = more pruning = simpler tree
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=1, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)

print(f"Built {len(clfs)} trees, one for each alpha value")
print(f"Number of nodes in the LAST tree (highest alpha): {clfs[-1].tree_.node_count}")
print(f"Number of nodes in the FIRST tree (alpha=0, no pruning): {clfs[0].tree_.node_count}")
```

```python
# ============================================================
# CELL 9c: REMOVING THE TRIVIAL LAST TREE
# ============================================================

# The very last tree (highest alpha) typically has just 1 node 
# (a "stump" that only predicts the majority class for everyone)
# We drop it since it's not a meaningful model to compare
clfs = clfs[:-1]
ccp_alphas = ccp_alphas[:-1]

print(f"Remaining trees to evaluate: {len(clfs)}")
```

```python
# ============================================================
# CELL 9d: NODES AND DEPTH VS ALPHA
# ============================================================

node_counts = [clf.tree_.node_count for clf in clfs]
depth = [clf.tree_.max_depth for clf in clfs]

fig, ax = plt.subplots(2, 1, figsize=(10, 7))
ax[0].plot(ccp_alphas, node_counts, marker="o", drawstyle="steps-post")
ax[0].set_xlabel("alpha")
ax[0].set_ylabel("number of nodes")
ax[0].set_title("Number of nodes vs alpha")

ax[1].plot(ccp_alphas, depth, marker="o", drawstyle="steps-post")
ax[1].set_xlabel("alpha")
ax[1].set_ylabel("depth of tree")
ax[1].set_title("Depth vs alpha")

fig.tight_layout()
plt.show()
```

#### Cost Complexity Pruning - Nodes and Depth vs Alpha
- Both node count and tree depth drop sharply as alpha increases from 
  0.000 to approximately 0.002-0.0025.
- This steep initial drop indicates that a large portion of the fully-grown 
  tree's complexity (many of its 97 nodes) was capturing noise rather than 
  meaningful patterns — these branches get pruned away with even a small 
  increase in alpha.
- Beyond this initial drop, the tree complexity decreases more gradually, 
  suggesting the remaining structure represents more genuine patterns 
  in the data.

```python
# ============================================================
# CELL 9e: RECALL VS ALPHA - TRAIN AND TEST
# ============================================================

recall_train = []
for clf in clfs:
    pred_train = clf.predict(X_train)
    recall_train.append(recall_score(y_train, pred_train))

recall_test = []
for clf in clfs:
    pred_test = clf.predict(X_test)
    recall_test.append(recall_score(y_test, pred_test))

fig, ax = plt.subplots(figsize=(15, 5))
ax.set_xlabel("alpha")
ax.set_ylabel("Recall")
ax.set_title("Recall vs alpha for training and testing sets")
ax.plot(ccp_alphas, recall_train, marker="o", label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, recall_test, marker="o", label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
# ============================================================
# CELL 9f: FINDING THE BEST ALPHA (HIGHEST TEST RECALL)
# ============================================================

# np.argmax finds the index of the highest value in the list
index_best_model = np.argmax(recall_test)
best_model = clfs[index_best_model]

print(f"Best alpha value: {ccp_alphas[index_best_model]:.5f}")
print(f"Test Recall at this alpha: {recall_test[index_best_model]:.4f}")
print(f"Number of nodes in this tree: {best_model.tree_.node_count}")
print(f"Depth of this tree: {best_model.tree_.max_depth}")
```

#### Note on Selecting Best Alpha

Selecting alpha purely by maximum test Recall chose alpha=0 (no pruning 
at all) — simply recreating our overfit 97-node baseline tree. This 
repeats the same flaw seen in pre-pruning: optimizing for Recall alone 
always favors maximum complexity.

**Revised approach:** We select alpha based on maximum test F1 score 
instead, which balances Recall and Precision and naturally favors a 
simpler, more generalizable tree over one that simply memorizes training data.

```python
# ============================================================
# CELL 9f (REVISED): FINDING THE BEST ALPHA USING F1 SCORE
# ============================================================

# Switching to F1 score - balances Recall and Precision, and naturally 
# favors simpler trees since overly complex trees tend to overfit 
# and hurt test Precision
f1_test = []
for clf in clfs:
    pred_test = clf.predict(X_test)
    f1_test.append(f1_score(y_test, pred_test))

index_best_model = np.argmax(f1_test)
best_model = clfs[index_best_model]

print(f"Best alpha value: {ccp_alphas[index_best_model]:.5f}")
print(f"Test F1 at this alpha: {f1_test[index_best_model]:.4f}")
print(f"Test Recall at this alpha: {recall_test[index_best_model]:.4f}")
print(f"Number of nodes in this tree: {best_model.tree_.node_count}")
print(f"Depth of this tree: {best_model.tree_.max_depth}")
```

```python
# ============================================================
# CELL 9g: F1 SCORE VS ALPHA - VISUALIZING THE TRADE-OFF
# ============================================================

fig, ax = plt.subplots(figsize=(15, 5))
ax.set_xlabel("alpha")
ax.set_ylabel("F1 Score")
ax.set_title("F1 Score vs alpha (Test Set)")
ax.plot(ccp_alphas, f1_test, marker="o", drawstyle="steps-post")
plt.show()
```

#### F1 Score vs Alpha - Key Finding

- F1 score remains relatively flat and high (~0.85-0.93) for alpha values 
  between 0 and approximately 0.02, then collapses sharply beyond that point.
- This wide plateau means many different alpha values achieve nearly 
  identical F1 performance — we are not forced to choose alpha=0 
  (the fully-grown, overfit 97-node tree) just because it is 
  mathematically the single highest point.
- We will select an alpha from within this plateau that meaningfully 
  reduces tree complexity while sacrificing minimal F1 performance, 
  rather than defaulting to the most complex tree.

```python
# ============================================================
# CELL 9h: SELECTING ALPHA FROM THE PLATEAU (BALANCING SIMPLICITY)
# ============================================================

# Instead of picking alpha=0, we look for an alpha that gives us 
# at least 95% of the maximum F1 score, but with fewer nodes (simpler tree)
max_f1 = max(f1_test)
threshold = 0.95 * max_f1  # we'll accept trees within 95% of best F1

# Find all alphas that meet this threshold
candidates = [(ccp_alphas[i], f1_test[i], clfs[i].tree_.node_count) 
              for i in range(len(clfs)) if f1_test[i] >= threshold]

# Among those candidates, pick the one with the FEWEST nodes (simplest tree)
candidates_sorted = sorted(candidates, key=lambda x: x[2])
best_alpha, best_f1, best_nodes = candidates_sorted[0]

print(f"Selected alpha: {best_alpha:.5f}")
print(f"F1 at this alpha: {best_f1:.4f}")
print(f"Number of nodes: {best_nodes}")
```

#### Selected Alpha for Post-Pruning

- Selected alpha = 0.00165, giving F1 = 0.891 (vs. maximum possible 0.930 at alpha=0).
- This alpha reduces the tree from 97 nodes down to just 17 nodes — 
  an 82% reduction in complexity for only a ~4% drop in F1 score.
- This is a far more interpretable and generalizable tree than the 
  fully-grown baseline, with minimal performance trade-off.

```python
# ============================================================
# CELL 9i: BUILDING THE FINAL POST-PRUNED MODEL
# ============================================================

estimator_2 = DecisionTreeClassifier(ccp_alpha=best_alpha, random_state=1)
estimator_2.fit(X_train, y_train)

print("Post-pruned model trained.")
print(f"Number of nodes: {estimator_2.tree_.node_count}")
print(f"Depth: {estimator_2.tree_.max_depth}")
```
